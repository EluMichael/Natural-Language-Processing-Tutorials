# Fourier Transform
The Fourier Transform is one of the most fundamental operations in applied mathematics and signal processing.

It transforms our time-domain signal into the frequency domain. Whereas the time domain expresses our signal as a sequence of samples, the frequency domain expresses our signal as a superposition of sinusoids of varying magnitudes, frequencies, and phase offsets.

```python
# Import the libraries
import librosa
import librosa.display
import scipy
import numpy as np
import IPython.display as ipd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the audio file
x, sr = librosa.load('data/english.wav')
```

```python
# Visualizing
plt.figure(figsize=(15, 5))
librosa.display.waveplot(x, sr, alpha=0.8)
ipd.Audio(x, rate=sr)
```

```python
# Compute the Fourier transform
X = scipy.fft(x)
X_mag = np.absolute(X)
f = np.linspace(0, sr, len(X_mag)) # frequency variable
```

```python
# Plot the spectrum
plt.figure(figsize=(13, 5))
plt.plot(f, X_mag) # magnitude spectrum
plt.xlabel('Frequency (Hz)')
```

```python
# Zoom in
plt.figure(figsize=(13, 5))
plt.plot(f[:5000], X_mag[:5000])
plt.xlabel('Frequency (Hz)')
```

# Feature Extraction

Somehow, we must extract the characteristics of our audio signal that are most relevant to the problem we are trying to solve. For example, if we want to classify instruments by timbre, we will want features that distinguish sounds by their timbre and not their pitch. If we want to perform pitch detection, we want features that distinguish pitch and not timbre.

This process is known as feature extraction.

### 1. Load the audio file

```python
# Import the libraries
%matplotlib inline
import numpy, scipy, matplotlib.pyplot as plt, IPython.display as ipd
import librosa, librosa.display
import stanford_mir
```

```python
# Initialize the stanford_mir
stanford_mir.init()
```

```python
# Load the audio file
x, sr = librosa.load('data/english.wav')
librosa.display.waveplot(x, sr, alpha=0.8)
ipd.Audio(x, rate=sr)
```

### 2. STFT (Short-Time Fourier Transform)
The short-time Fourier transform (STFT) is obtained by computing the Fourier transform for successive frames in a signal.

```python
# Computes the STFT (Short-Time Fourier Transform)
hop_length = 512                                    # hop length (i.e. the frame increment)
n_fft = 2048                                        # the size of the FFT
X = librosa.stft(x, n_fft = n_fft, hop_length=hop_length)               # STFT
```

```python
# Convert the hop length and frame size to units of seconds
print(float(hop_length)/sr)
print(float(n_fft)/sr)
```

For real-valued signals, the Fourier transform is symmetric about the midpoint. Therefore, librosa.stft only retains one half of the output:

```python
X.shape
```

This STFT has 1025 frequency bins and 9813 frames in time.

### 3. Spectrogram
The spectrogram shows the the intensity of frequencies over time. A spectrogram is simply the squared magnitude of the STFT.

```python
# Our perception of sound intensity is logarithmic in nature. So, we are often interested in the log amplitude
S = librosa.amplitude_to_db(abs(X))
```

```python
# Visualize the spectogram
plt.figure(figsize=(15, 5))
plt.set_cmap('hot')
librosa.display.specshow(S, sr=sr, hop_length=hop_length, x_axis='time', y_axis='linear')
plt.colorbar(format='%+2.0f dB')
```

### 4. Mel-spectrogram

```python
# Compute the Mel-spectrogram
hop_length = 256
S = librosa.feature.melspectrogram(x, sr=sr, n_fft=4096, hop_length=hop_length)
```

```python
# The human perception of sound intensity is logarithmic in nature. Therefore, like the STFT-based 
# spectrogram, we are often interested in the log amplitude
logS = librosa.power_to_db(abs(S))
```

```python
# Visualize the  Mel-spectrogram 
plt.figure(figsize=(15, 5))
plt.set_cmap('hot')
librosa.display.specshow(logS, sr=sr, hop_length=hop_length, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
```

### 5. CQT (Constant-Q Transform)
Unlike the Fourier transform, but similar to the mel scale, the constant-Q transform uses a logarithmically spaced frequency axis.

```python
# Compute the CQT
fmin = librosa.midi_to_hz(36)
C = librosa.cqt(x, sr=sr, fmin=fmin, n_bins=72)
logC = librosa.amplitude_to_db(abs(C))
```

```python
# Viualize the CQT
plt.figure(figsize=(15, 5))
librosa.display.specshow(logC, sr=sr, x_axis='time', y_axis='cqt_note', fmin=fmin, cmap='coolwarm')
plt.colorbar(format='%+2.0f dB')
```

### 6. Mel Frequency Cepstral Coefficients (MFCCs)
The mel frequency cepstral coefficients (MFCCs) of a signal are a small set of features (usually about 10-20) which concisely describe the overall shape of a spectral envelope. In MIR, it is often used to describe timbre.

```python
# Import the libraries
import numpy, scipy, sklearn, urllib, IPython.display
import matplotlib.pyplot as plt
import librosa, librosa.display
plt.rcParams['figure.figsize'] = (14,4)
```

```python
# Loading the audio file
x, fs = librosa.load('./data/english.wav')
librosa.display.waveplot(x, sr=fs)
IPython.display.Audio(x, rate=fs)
```

```python
# Compute the MFCC
mfccs = librosa.feature.mfcc(x, sr=fs)
print(mfccs.shape)
```

```python
# Visualize the MFCC
librosa.display.specshow(mfccs, sr=fs, x_axis='time')
```

```python
# Scale the MFCCs such that each coefficient dimension has zero mean and unit variance
mfccs = sklearn.preprocessing.scale(mfccs, axis=1)
print(mfccs.mean(axis=1))
print(mfccs.var(axis=1))
```

```python
# Visualize the MFCC
librosa.display.specshow(mfccs, sr=fs, x_axis='time')
```
